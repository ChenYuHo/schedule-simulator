# Explore the relationship between batch size and execution time

```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
def plot(report_path):
    with open("batch_size_analysis_reports/{}".format(report_path)) as file:
        report = json.load(file)
    for k, v in report["args"].items():
        print("{}: {}".format(k, v))
    print("Max batch size on host {}: {}".format(report["host"],
                                                 max([int(x) for x in report["timings"].keys()])))
    x = list()
    y = list()
    for k, v in report["timings"].items():
        x.append(int(k))
        y.append(min(v)/1e6)
    print(np.polyfit(x, y, 1))
    plt.figure(figsize=(14,8))
    plt.plot(x,y, "-o")
    plt.xlabel("Batch size")
    plt.ylabel("Execution time per batch (ms)")
    plt.grid()
    plt.show()
```

```python
plot("VGG16_08-17-10-38.batch_size_analysis.json")
```
